```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
covid_df = pd.read_csv('owid-covid-data.csv')
```

# Data wrangling

```python
covid_df = covid_df.drop(columns=(['handwashing_facilities', 'new_people_vaccinated_smoothed',
       'new_people_vaccinated_smoothed_per_hundred', 'stringency_index',
       'population', 'population_density', 'median_age', 'aged_65_older',
       'aged_70_older', 'gdp_per_capita', 'extreme_poverty',
       'cardiovasc_death_rate', 'diabetes_prevalence', 'female_smokers',
       'male_smokers', 'hospital_beds_per_thousand', 'life_expectancy',
       'human_development_index', 'excess_mortality_cumulative_absolute',
       'excess_mortality_cumulative', 'excess_mortality',
       'excess_mortality_cumulative_per_million']))
countries =['Europe', 'European Union', 'Africa', 'Asia', 'International', 'Low income', 'Lower middle income', 'North America', 'Oceania', 'South Africa', 'South America', 'South Korea', 'Upper middle income', 'World']
covid_df = covid_df[np.isin(covid_df['location'], countries, invert=True)]
covid_df['date'] = covid_df['date'].astype('datetime64[ns]')
```

# Data Analysis and Vizualization

```python
def coutries_graphs(country):

    '''Funcion para graficar distintos features de un pais determinado.

        Function to graph different features of a given country.
    
    Parameters:
        country: es un String.

        country: is a String.
    
    Return:
        Retorna una figura con 5 graficos relacionados al mismo pais.

        Returns a figure with 5 graphs related to the same country.
    '''

    data_df = covid_df[covid_df['location']==country]
    fig, (ax1, ax2, ax3, ax4 ,ax5) = plt.subplots(5,1, constrained_layout=True,sharex=True)
    fig.set_size_inches(12, 12, forward=True)
    fig.suptitle('Reporte Covid-19 de {}.'.format(country), fontsize=18)
    ax1.plot(data_df['date'], data_df[['new_cases']], c='r')
    ax1.set_ylabel('casos nuevos.')
    ax1.set_title('Casos Diarios.')
    ax1.grid()
    ax2.plot(data_df['date'], data_df[['new_deaths']], c='b')
    ax2.set_ylabel('muertos nuevos.')
    ax2.set_title('Muerto Diarios.')
    ax2.grid()
    ax3.plot(data_df['date'], data_df[['positive_rate']], c='g')
    ax3.set_ylabel('positividad.')
    ax3.set_title('Positividad.')
    ax3.grid()
    ax4.plot(data_df['date'], data_df[['icu_patients']], c='black')
    ax4.set_ylabel('pacientes uci.')
    ax4.set_title('Pacientes UCI.')
    ax4.grid()
    ax5.plot(data_df['date'], data_df[['new_vaccinations']], c='y')
    ax5.set_ylabel('vacunados diarios.')
    ax5.set_xlabel('Fecha de registro.')
    ax5.set_title('Vacunacion.')
    ax5.grid()
    plt.show()
```

En esta figura se muentran 5 graficos, donde se ven los siguientes features: casos nuevos, fallecidos nuevos, positividad, pacientes UCI y vacunados, en una linea dea tiempo desde el inicio de la pandemia hasta la actualidad.

This figure shows 5 graphs, where the following features are seen: new cases, new deaths, positivity, ICU and vaccinated patients, in a timeline from the start of the pandemic to the present.

```python
coutries_graphs('Chile')
```

Se realiza mapa de calor solo para ver la relacion entre algunos features.

A heat map is made only to see the relationship between some features.

```python
def heatmap_graph(country):
    '''Funcion muestra la correlacion en forma de mapa de calor de un determinado pais.

     Function displays the correlation in the form of a heat map for a given country.
    
    Parameters:
        country: es un String.

        country: is a String.
    
    Return:
        Retorna un mapa de calor.

        Return a heat map.
    '''
    data_df = covid_df[covid_df['location']==country]
    heat = data_df[['new_cases', 'new_deaths', 'new_vaccinations', 'new_cases_per_million', 'new_deaths_per_million', 'new_tests_per_thousand','positive_rate','icu_patients',  'total_vaccinations']].corr()
    plt.figure(figsize=(12,12))
    sns.heatmap(heat, annot=True, cmap='YlGnBu')
    plt.title('Mapa de calor de {}'.format(country))
    plt.show()
```

En esta figura se muentra un grafico de barra que hace la comparacion entre dos paises, tomando en cuenta un feature en particular.

This figure shows a bar graph that makes the comparison between two countries, taking into account a particular feature.

```python
def country_vs_country(country1, country2, feature):
    
    '''Funcion para graficar la compracion entre dos paises dado un feature dado.

        Function to graph the purchase between two countries given a given feature.
    
    Parameters:
        country1: es un String.
        country2: es un String.
        feature: es un String.

        country1: is a String.
        country2: is a String.
        feature: is a String.
    
    Return:
        Retorna una figura con un grafico de barra comprando dos paises.

        Returns a figure with a bar graph buying two countries.
    '''
    df_1 = covid_df[covid_df['location']==country1]
    df_1_y = df_1[feature].max()
    df_2 = covid_df[covid_df['location']==country2]
    df_2_y = df_2[feature].max()
    fig, ax = plt.subplots(constrained_layout=True)
    ax.bar(country1, df_1_y, width=0.5, label=country1)
    ax.bar(country2, df_2_y, width=0.5, label=country2)
    ax.set_ylabel('Cantidad de casos.')
    ax.set_xlabel('Paises.')
    ax.set_title('Comparacion de {} entre {} y {}.'.format(feature,country1,country2))
    plt.legend()
    plt.show()
```

Para hacer mas interactivo y atractivo visualmente para el usuario, se usara la libreria Plotly.

To make it more interactive and visually attractive for the user, the Plotly library will be used.

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px
```

```python
def graph_plot(country):

    '''Funcion para graficar distintos features de un pais determinado.

        Function to graph different features of a given country.
    
    Parameters:
        country: es un Dataframe.

        country: is a Dataframe.
    
    Return:
        Retorna una figura con 5 graficos relacionados al mismo pais.

        Returns a figure with 5 graphs related to the same country.
    '''
    data_df = df_final[df_final['location']==country]  
    fig = make_subplots(rows=5, cols=1,
                            shared_xaxes=True,
                            vertical_spacing=0.02)
    fig.add_trace(
        go.Scatter(x=data_df['date'], y=data_df['new_cases'],name=('Casos nuevos')),
        row=1, col=1
    )
    fig.add_trace(
        go.Scatter(x=data_df['date'], y=data_df['new_deaths'],name=('Fallecidos nuevos')),
        row=2, col=1
    )
    fig.add_trace(
        go.Scatter(x=data_df['date'], y=data_df['positive_rate'],name=('Positividad')), 
        row=3, col=1
    )
    fig.add_trace(
        go.Scatter(x=data_df['date'], y=data_df['icu_patients'],name=('Pacientes UCI')),
        row=4, col=1
    )
    fig.add_trace(
        go.Scatter(x=data_df['date'], y=data_df['new_vaccinations'],name=('Vacunados')),
        row=5, col=1
    )
    fig.update_xaxes(
        rangeselector=dict(
            buttons=list([
                dict(count=1, label="1m", step="month", stepmode="backward"),
                dict(count=6, label="6m", step="month", stepmode="backward"),
                dict(count=1, label="1y", step="year", stepmode="backward"),
                dict(step="all")
            ])
        )
    )
    fig.update_layout(height=800, width=800,
                        title_text="Reporte Covid-19 de {}".format(country), autosize=False)
    fig.show(config= {'displaylogo': False})
```

```python
graph_plot('Chile')
```

```python
def bar_plot(country1, country2, feature):
    '''
    Funcion para graficar la compracion entre dos paises dado un feature dado.
    Function to graph the purchase between two countries given a given feature.
    
    Parameters:
        country1: es un String.
        country2: es un String.
        feature: es un String.

        country1: is a String.
        country2: is a String.
        feature: is a String.
    
    Return:
        Retorna una figura con un grafico de barra comparando dos paises.

        Returns a figure with a bar graph comparing two countries.
    '''
    x = [feature]
    df1 = df_final[df_final['location']==country1]
    df2 = df_final[df_final['location']==country2]
    y1 = df1[feature].max()
    y2 = df2[feature].max()
    fig = go.Figure(data=[
            go.Bar(name=country1, x=x, y=[y1], text=y1, textposition='auto'),
            go.Bar(name=country2, x=x, y=[y2], text=y2, textposition='auto')
        ])
    fig.update_layout(title='Comparacion de {} entre {} y {}'.format(feature, country1, country2), yaxis=dict(
         title='Numero de casos',
         titlefont_size=16,
         tickfont_size=14,
     ),barmode='group',bargap=0.5, autosize=False)
    fig.show(config= {'displaylogo': False})
```

```python
bar_plot('Chile', 'Peru', 'total_deaths_per_million')
```

```python
def heat_plotly(country):   
    '''
    Funcion muestra la correlacion en forma de mapa de calor de un determinado pais.
    Function displays the correlation in the form of a heat map for a given country.
    
    Parameters:
        pais: es un String.
        pais: is a String.
    
    Return:
        Retorna un mapa de calor.
        Return a heat map.
    '''
    data_df = df_final[df_final['location']==country]
    heat = data_df[['new_cases', 'new_deaths', 'new_vaccinations', 'new_cases_per_million', 'new_deaths_per_million', 'new_tests_per_thousand','positive_rate','icu_patients',  'total_vaccinations']].corr()
    fig = px.imshow(heat,  width=800, height=800, aspect='auto', text_auto='.2f')
    fig.update_layout(title='Correlacion de caracteristicas de pais' ,autosize=False)
    fig.show(config= {'displaylogo': False})
```

```python
def mapplotly(feature):
    '''Funcion muestra una figura del mapa del mundo.

        Function shows a figure of the world map.
    
    Parameters:
        feature: es un String.

        feature: is a String.
    
    Return:
        Retorna un mapa del mundo comprando todods los paises dado un feature de entrada.

        Returns a map of the world buying all the countries given an input feature.
    '''
    dff = df_final.groupby(['location','iso_code','continent']).agg({'total_deaths_per_million':'max', 'total_cases_per_million':'max', 'total_vaccinations_per_hundred':'max','total_vaccinations':'max'}).reset_index()
    fig = go.Figure(data=go.Choropleth(
        locations = dff['iso_code'],
        z = dff[feature].round(decimals=2),
        text = dff['location'],
        colorscale = 'Reds',
        autocolorscale=False,
        reversescale=True,
        marker_line_color='darkgray',
        marker_line_width=0.0,
        colorbar_title = feature,
    ))
    fig.update_layout(title='Mapa Mundial' ,autosize=False)

    fig.show(config= {'displaylogo': False, 'displayModeBar': True,'modeBarButtonsToRemove': ['zoom','Pan', 'Lasso', 'select']})
```
